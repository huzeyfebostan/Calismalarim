# Karşılaştırma Grafikleri

Karşılaştırmalı grafikleri, şu ana kadar öğrendiğimiz grafiklerin 2D versiyonlarıdır. Burada tartışılan iki temel çizim türü:
- **jointplot()**
  - Her bir özellik içindeki dağılımları netleştirmek için her özelliğin dağılım grafiği ile histogramını eşler.
  - Ayrıca dağılım grafiğini altıgen (hexogonal) bir çizim veya 2D KDE grafiği olacak şekilde ayarlayabilir.

- **pairplot()**
  - pairplot(), bir DataFrame'deki tüm sayısal sütunları karşılaştırmanın hızlı bir yoludur. 
  - Her sütun için otomatik olarak bir histogram ve olası tüm sütun kombinasyonları arasında bir dağılım grafiği karşılaştırması oluşturur.
  - pairplot(), çok fazla sütunlu büyük DataFrame'ler için yoğun CPU ve RAM kullanabilir.
  - Yalnızca ilgilendiğiniz sütunlara filtrelemek iyi bir fikirdir.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Veri

```python
df = pd.read_csv("StudentsPerformance.csv")
```

```python
df.head()
```

## jointplot()

```python
# kind=scatter varsayılan
sns.jointplot(x='math score',y='reading score',data=df)
```

```python
# Yüzeyine daha fazla nokta düşen Hexagons karanlıktır
sns.jointplot(x='math score',y='reading score',data=df,kind='hex')
```

```python
sns.jointplot(x='math score',y='reading score',data=df,kind='kde')
```

## pairplot()

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='gender',palette='viridis')
```

```python
sns.pairplot(df,hue='gender',palette='viridis',corner=True)
```

```python
sns.pairplot(df,hue='gender',palette='viridis',diag_kind='hist')
```
