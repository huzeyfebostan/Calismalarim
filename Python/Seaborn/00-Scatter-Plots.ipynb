# scatterplot()
Verinizdeki iki sürekli özellik arasındaki ilişkiyi kolayca görebileceğiniz grafik türüdür.
- Sürekli özellik örnekleri:
       - Yaş
       - Yükseklik
       - Maaş
       - Sıcaklık
       - Fiyatlar

```python
# Çalışan maaşları ile satış tutarı arasındaki ilişki 
sns.scatterplot(x='salary',y='sales',data=df)
```

## Matplotlib'deki Figüre Nesenesini Kullanma
**matplotlib'in ile seaborn birbirine bağlı olarak kullanılabilir.  matplotlib.pyplot'u içe aktarır ve seaborndaki figürü doğrudan etkilemek için matplotlib çağrıları yapabiliriz.**

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8,4),dpi=200)
sns.scatterplot(x='salary',y='sales',data=df);
```

## Seaborn Parametereleri

**Hue** ve **palette** parametreleri, seaborn'daki birçok çizim çağrısında yaygın olarak bulunur.

### hue Parametresi

Dağılım grafiklerinde, DataFrame'deki kategorik bir özelliği temel alan renklendirme oluşturur.

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x='salary',y='sales',data=df,hue='division')
```

```python
df['work experience'].sort_values().unique()
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x='salary',y='sales',data=df,hue='work experience');
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x='salary',y='sales',data=df,
                hue='work experience',
                palette='viridis')
```

## scatterplot 'un Parametreleri

Bu parametreler daha çok scatterplot() çağrısına özeldir.

### Size parametresi

Başka bir sütuna göre veri noktalarını (işaretçileri) boyutlandırmanıza izin verir.

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x='salary',y='sales',data=df,hue='work experience',size='work experience')
```

İşaretçinin boyutunu kontrol edin ve her benzersiz değerin göstergede (legend) görünmesi için lengend='full' olarak ayarlayın:

```python
plt.figure(figsize=(12,8))
sns.scatterplot(data=df, x="salary", y="sales", 
                size="work experience", sizes=(20, 200), legend="full")
```

**İşaretçi boyutunu tek tip kullanmak için *s* veya *size* parametresi**

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x='salary',y='sales',data=df,s=200)
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x='salary',y='sales',data=df,s=200,linewidth=0,alpha=0.2)
```

plt.figure(figsize=(12,8))

### Style parametresi

Veri kümesindeki başka bir kategorik özelliğe dayalı olarak stilleri otomatik olarak seçmeyi sağlar. İsteğe bağlı olarak, bir işaretçi listesi vermek için **markers=** parametresini bir işaretci listesi ile kullanılır, örneğin: 
[',', '.', 'o', 'v', '^', '<', '>', '8', 's', 'p', '*', 'h', 'H', 'D', 'd', 'P', 'X']

```python
df['level of education'].unique()
```

```python
markers = {"some college": "s", "associate's degree": "X", "high school": "p",
           "bachelor's degree":"o","master's degree":"8"}

plt.figure(figsize=(12,8))
sns.scatterplot(x='salary',y='sales',data=df,style='level of education',
                markers=markers,s=200,linewidth=0,alpha=0.2)
```

```python
plt.figure(figsize=(12,8))
# Bazen aynı sütunda hem hue hem de stil ayarlamak güzel bir görünüm verir
sns.scatterplot(x='salary',y='sales',data=df,style='level of education',
                hue='level of education',s=100)
```

## Seaborn Figure'ü Dışa Aktarma

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x='salary',y='sales',data=df,style='level of education',
                hue='level of education',s=100)

# Aynı hücrede savefig'i çağırın
plt.savefig('example_scatter.jpg')
```
