#### Yeni Sütunlar Oluşturma
DataFrame sütunları üzerinde Numpy dizilerinde gerçekleştirdiğimiz işlemlere benzer işlemler gerçekleştirebiliriz. Bu işlemler sonucunda da yeni sütunlar oluşturabiliriz.

Örneğin; toplam faturaya bahşişin yüzde olarak oranını **tip_percentage** olarak yeni bir sütun oluşturalım.

```python
df['tip_percentage'] = 100* df['tip'] / df['total_bill']
```

```python
df.head()
```

```python
# partideki kişi başıca düşen ödeme miktarını veren yeni bir sütun
df['price_per_person'] = df['total_bill'] / df['size']
```

```python
df.head()
```

```python
# Yeni sütunda gözüken ondalıklı sayıları yuvarlamak istersek  
# Numpy fonksiyonu np.round kullanabiliriz.
help(np.round)
```

#### Mevcut Sütunları Ayarlama

```python
df['price_per_person'] = np.round(df['price_per_person'],2)
```

```python
df.head()
```

#### Sütunları Silme
Dikkat etmemiz gereken şey sütunu sileceğimizi belirtmek için **axis=1** seçeneğini eklemeliyiz. Kalıcı olarak silmek için ya **inplace** seçeneğini eklemeliyiz veya daha kolayca **df =** olarak drop etmeliyiz.

```python
df = df.drop("tip_percentage",axis=1)
```

```python
df.head()
```

```python
df.shape
```

## Satırlar
#### Tek Satıra Ait Bilgileri Alma
Bunun için tamsayı tabanlı indeks için **.iloc** veya etiket (veya isim) tabanlı indeks için **.loc** yöntemlerini kullanılır

```python
# DataFrame'de Tamsayı Tabanlı bir indeks konumuna bakma
df.iloc[0] # 0 indeksli satırdaki bilgileri çeker.
```

```python
# DataFrame'de etiket veya isim Tabanlı bir indeks konumuna bakma
df.loc['Sun2959'] # Sun2959 indeksli satırdaki bilgileri çekiyoruz.
```

#### Çoklu Satırlara Ait Bilgileri Alma
iloc 'a dayalı olarak birden fazla satır almak istersek, Python'un dilimleme gösterilimi kullanılabilir.

```python
df.iloc[0:4]
```

loc 'a tabanlı birden fazla satıra ait bilgileri almak istersek, liste parantezi içerisinde bir isim veya etiket string kümesi girmeliyiz.

```python
df.loc[['Sun2959','Sun5260']]
```

#### Satır Silme

Veri kümeleri, belirli bir koşula karşı gelen satır konumlarını bilemeyeceğimizden böyle satırlar kaldırmayacağımız kadar büyük olacaktır. Bunun yerine satırları eksik veri veya sütun değerleri gibi koşullara göre atarız.

```python
df.head()
```

```python
df.drop('Sun2959',axis=0) # Sun2959 etiketli satırı silme
```

```python
df.head()
```

#### Yeni Satır Ekleme

Aşağıdaki gibi tek bir satır eklemek oldukça nadirdir. Genellikle aynı anda birden fazla satır eklemek için **.concat()** kullanılır. **.append()** yöntemini pd.Series() nesnesine ait listeyle kullanabilirsiniz.

```python
one_row = df.iloc[0]
```

```python
one_row
```

```python
type(one_row)
```

```python
df.tail()
```

```python
df.append(one_row)
```

```python
# pd.concat ile bir satır Series ekleme kodu:
pd.concat([df,one_row.to_frame().T])
```
