** Bir pivot ne tür bir soruyu yanıtlamaya yardımcı olur?**

**Google'ın her bir ürün türünden kaç lisans satın aldığını bilmek istediğimizi hayal edin. Şu anda verilerin biçimlendirilme şeklinin okunması zor. Bunu daha açık hale getirmek için divotlayalım, elimizdeki soru için verilerin bir alt kümesini alacağız.**

```python
# Bir alt küme alalım, aksi takdirde yinelenen satırlar ve 
# veriler nedeniyle bir hata alırız.
licenses = df[['Company','Product','Licenses']]
licenses
```

```python
pd.pivot(data=licenses,index='Company',columns='Product',values='Licenses')
```

Bunu çalıştırdığınızda bir sürü NaN değeri gözükür. Bu da mantıklı, çünkü her ürünü her şirkete satmadık. Şimdi Google'ı yakalamak için indekslemeyi kullanabiliriz. Bunun üzerine bir toplama çağrısı gerçeklemek isteyelim.

## pivot_table() metodu

pivot() yöntemine benzer şekilde, pivot_table(), bir pivot çağrısına toplama işlevleri ekleyebilir. Bunun için DataFrame'mimizi pivot tablosuna aktaracağız ve ardından yeni indeksin ne olacağını seçmemiz gerekiyor, bu aslında neye göre gruplayacağınızıdır. Ve son olarak, gerçekten kullanmak istediğiniz toplama fonksiyonu nedir?

Şu soruyu yanıtlamaya çalışıyoruz: Her bir şirkete gerçekte ne kadar sattım?

```python
df
```

```python
# Account Number sum() tutmak/kullanmak mantıklı 
# olmadığına dikkat edin.
pd.pivot_table(df,index="Company",aggfunc='sum')
```

Bu kod buradaki her şirketin hesap numaralarını da toplar ki Google için tüm hesap numaralarını toplamak mantıklı değildir, çünkü biraz anlamsızdır. Ancak satılan toplam lisansları toplam satış fiyatında toplamak mantıklıdır.Bu tablodan Google'ın 600 lisansı toplam 3150000$ sattığı görülür.

```python
df.groupby('Company').sum() # bunu kolayca groupby ile yaparız.
```

```python
# İstediğiniz sütunları alabilirsiniz
pd.pivot_table(df,index="Company",aggfunc='sum')[['Licenses','Sale Price']]
```

```python
# Veya bunları istenen değerler olarak belirtin
pd.pivot_table(df,index="Company",aggfunc='sum',values=['Licenses','Sale Price'])
```

```python
df.groupby('Company').sum()[['Licenses','Sale Price']]
```

```python
# İki sütunlu indekse sahipsek
pd.pivot_table(df,index=["Account Manager","Contact"],values=['Sale Price'],aggfunc='sum')
```

Sütunlar isteğe bağlıdır - önemsediğiniz gerçek değerleri bölümlere ayırmanın ek bir yolunu sağlarlar. Toplama işlevleri, listelediğiniz değerlere uygulanır.

```python
pd.pivot_table(df,index=["Account Manager","Contact"],values=["Sale Price"],columns=["Product"],aggfunc=[np.sum])
```

```python
pd.pivot_table(df,index=["Account Manager","Contact"],values=["Sale Price"],columns=["Product"],aggfunc=[np.sum],fill_value=0)
```

```python
# Birden çok agg fonksiyonu ekleyebilir
pd.pivot_table(df,index=["Account Manager","Contact"],values=["Sale Price"],columns=["Product"],
               aggfunc=[np.sum,np.mean],fill_value=0)
```

```python
# Birden çok sütuna ekleyebilir
pd.pivot_table(df,index=["Account Manager","Contact"],values=["Sale Price","Licenses"],columns=["Product"],
               aggfunc=[np.sum],fill_value=0)
```

```python
# Birden çok sütuna ekleyebilir
pd.pivot_table(df,index=["Account Manager","Contact","Product"],values=["Sale Price","Licenses"],
               aggfunc=[np.sum],fill_value=0)
```

```python
# get Final "ALL" with margins = True
# Can add on multiple columns
pd.pivot_table(df,index=["Account Manager","Contact","Product"],values=["Sale Price","Licenses"],
               aggfunc=[np.sum],fill_value=0,margins=True)
```

```python
pd.pivot_table(df,index=["Account Manager","Status"],values=["Sale Price"],
               aggfunc=[np.sum],fill_value=0,margins=True)
```
