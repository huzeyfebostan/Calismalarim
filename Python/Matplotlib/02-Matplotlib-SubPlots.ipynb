# plt.subplots()

**NOT: notebook'ta komutları aynı hücrede yazdığınızdan emin olun!**

plt.subplots() nesnesi birden fazla grafiği yan yana gösteren bir eksen yöneticisidir. figure ve axes nesnelerini içeren sayısal bir tuple dizisi döndürür.

```python
# fig ve eksen nesnelerini içeren tuple hariç 
# plt.figure() 'e benzerdir
fig, axes = plt.subplots()

# çizime özellikler eklemek için axes nesnesini kullanma
axes.plot(x,y,'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title'); # ; Out[]'u gizler
```

```python
# axes dizisini kullanarak çizim yapma
fig, axes = plt.subplots(nrows=1,ncols=2)

axes[0].plot(a,b)
axes[1].plot(x,y);
```

## subplots() nesnesinin parametreleri 

Hem figure hem de axes nesnelerine sahip olduğumuzdan parametreleri her iki seviyede de düzenleyebiliriz.

```python
fig, axes = plt.subplots(nrows=2,ncols=2,figsize=(8,4),dpi=200)

# ÖNCE EKSEN PARAMETRELERİNİ AYARLAYIN

# Eksen seviyesindeki parametreler
axes[0,0].plot(a,b)
axes[0,0].set_title('0 0 Title')


axes[1,1].plot(x,y)
axes[1,1].set_title('1 1 Title')
axes[1,1].set_xlabel('1 1 X Label')

axes[0,1].plot(y,x)
axes[0,1].set_title('0 1 Title')
axes[1,0].plot(b,a)
axes[1,0].set_title('1 0 Title')

# SONRA GENEL FIGURE PARAMETRELERİNİ AYARLAYIN

# Figure düzeyindeki parametreler
fig.suptitle("Figure Level",fontsize=20);
plt.tight_layout()
```

### subplots() boşluklarını elle ayarlama

Boşlukları manuel olarak **.subplots_adjust()** ile ayarlama

Örnek değerler:

- **left   = 0.125  :** eksenlerin sol tarafını, 
- **right  = 0.9    :** eksenlerin sağ tarafını, 
- **bottom = 0.1    :** eksenlerin alt tarafını ve 
- **top    = 0.9    :** eksenlerin üst tarafını genişletir (veya daraltır).
- **wspace = 0.2    :** ortalama eksen genişliğinin bir oranı ile verilen, çizimler arasındaki yatay boşluk miktarı
- **hspace = 0.2    :** ortalama eksen yüksekliğinin bir oranı ile verilen, çizimler arasındaki düşey boşluk miktarı

```python
fig,axes = plt.subplots(nrows=2,ncols=2,figsize=(12,8))

# Eksen seviyesindeki parametreler
axes[0,0].plot(a,b)
axes[1,1].plot(x,y)
axes[0,1].plot(y,x)
axes[1,0].plot(b,a)

# eksenleri genişletmek için sol, sağ, üst, alt boşluklar kullanılır
# çizimler arasına yatay veya düşey boşluk eklemek için wspace, hspace kullanılır

fig.subplots_adjust(left=.1,bottom=.1,right=.8,top=.5,
                    wspace=0.53,hspace=0.9)
```
