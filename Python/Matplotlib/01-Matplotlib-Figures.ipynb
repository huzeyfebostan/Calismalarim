## Figure Nesnesi Oluşturma

Nesne Yönelimli yöntemi kullanmanın ana fikri, figure nesnesi oluşturmak ve ardından bu nesneden yöntemleri veya nitelikleri çağırmaktır. Bu yaklaşım, üzerinde birden fazla çizim bulunan bir tuvalle uğraşırken oldukça iyidir.

**Figure nesnesi, ona eksen ekleyene kadar teknik olarak görünmez.**

### Figure'e başka bir eksen seti ekleme

Bu metodoloji, birden fazla eksen eklemenize, eksenleri hareket ettirmenize veya yeniden boyutlandırmanıza olanak verir.

Kod biraz karmaşıktır, ancak avantajı çizim eksenlerinin nereye yerleştirildiği konusunda tam kontrole sahip olmamız ve figure'e kolayca birden fazla eksen ekleyebilmemizdir.

```python
# Figure Nesnesi oluşturma
fig = plt.figure()

axes1 = fig.add_axes([0, 0, 1, 1]) # Daha büyük tuval

axes2 = fig.add_axes([0.2, 0.2, 0.5, 0.5]) # Daha küçük tuval

# Daha büyük Figure'ü Axes1 tuvaline çizme
axes1.plot(a,b)

# Axes1 figure'ne eklemek için set_ kullanma
axes1.set_xlabel('X Label')
axes1.set_ylabel('Y Label')
axes1.set_title('Big Figure')

# Figure'ü Axes2'ye yerleştirme
axes2.plot(a,b)
axes2.set_title('Small Figure')
axes2.set_xlim(6,8)
axes2.set_ylim(2000,4000);
```

Küçük şekli hareket ettirelim ve parametrelerini düzenleyelim

```python
# Boş kanvas oluşturma
fig = plt.figure()

axes1 = fig.add_axes([0, 0, 1, 1]) # Daha büyük figure

axes2 = fig.add_axes([0.2, 0.5, 0.25, 0.25]) # Daha küçük figure

# Daha büyük Figure'ü Axes 1'e yerleştirme
axes1.plot(a, b)

# Eksen figure'ne eklemek için set_ kullanma
axes1.set_xlabel('X Label')
axes1.set_ylabel('Y Label')
axes1.set_title('Big Figure')

# Figure'ü Axes 2'ye yerleştirme
axes2.plot(a,b)
axes2.set_xlim(8,10)
axes2.set_ylim(4000,10000)
axes2.set_xlabel('X')
axes2.set_ylabel('Y')
axes2.set_title('Zoomed In');
```

Uzunluk ve genişlik uygun verilmişse, ana figure'ün dışına bile istediğiniz kadar eksen ekleyebiliriz.

```python
# Boş kanvas oluşturma
fig = plt.figure()

axes1 = fig.add_axes([0, 0, 1, 1]) # Ana figure
axes2 = fig.add_axes([0.2, 0.5, 0.25, 0.25]) # Daha küçük figure
axes3 = fig.add_axes([1, 1, 0.25, 0.25]) 
# Sağ üst köşeden başlayan figure

# Daha büyük Figure'ü Axes1'e yerleştirme
axes1.plot(a, b)

# Eksen figure'ne eklemek için set_ kullanma
axes1.set_xlabel('X Label')
axes1.set_ylabel('Y Label')
axes1.set_title('Big Figure')

# Figure'ü Axes2'ye yerleştirme
axes2.plot(a,b)
axes2.set_xlim(8,10)
axes2.set_ylim(4000,10000)
axes2.set_xlabel('X')
axes2.set_ylabel('Y')
axes2.set_title('Zoomed In')

# Figure'ü Axes3'e yerleştirme
axes3.plot(a,b);
```

### Figure Parametereleri
**figsize** parametresi (width,height) tuple alır ve figure tuvalinin genişliğini ve yüksekliğini tanımlar. **dpi** parametresi ile resmin çıktı kalitesini belirleyebiliriz.

```python
# Boş kanvas 
fig = plt.figure(figsize=(10,4),dpi=300)

axes1 = fig.add_axes([0, 0, 1, 1])

axes1.plot(a,b);
```
