```python
import matplotlib.pyplot as plt
```

### Veri

```python
import numpy as np
```

```python
x = np.arange(0,10)
y = 2 * x
```

## Legends (Göstergeler)

Aynı figure'e birden fazla grafik eklendiğinde **label='label text'** argümanı kullanılarak figure'e eklemek için legend yöntemini kullanabilirsiniz:

```python
fig = plt.figure(figsize=(6,4))

ax = fig.add_axes([0,0,1,1])

ax.plot(x, x**2, label="x**2")
ax.plot(x, x**3, label="x**3")
ax.legend()
```

```python
fig, ax = plt.subplots(figsize=(6,4))

ax.plot(x, x**2, label="x**2")
ax.plot(x, x**3, label="x**3")
ax.legend()
```

**legend** fonksiyonu, göstergenin grafikte nerede konumlanacağını belirten isteğe bağlı bir **loc** argümanı alır. İzin verilen **loc** değerleri, göstergenin konumlanacağı çeşitli yerler için sayısal kodlardır.

```python
# Birçok seçenek....

ax.legend(loc=0) # en iyi konum
ax.legend(loc=1) # sağ üst köşe
ax.legend(loc=2) # sol üst köşe
ax.legend(loc=3) # sol alt köşe
ax.legend(loc=4) # sag alt köşe

# ... daha birçok seçenek mevcuttur

# en yaygın seçim
ax.legend(loc=0) # en uygun konuma yerleştirir
fig
```

```python
ax.legend(loc=(1.1,0.5)) 
# konumu manuel olarak bir tuple ile ayarlayabiliriz
fig
```

## Renk, çizgi genişliğini ve tipini ayarlama


### MatLab benzeri sözdizimi ile renkler

Matplotlib ile çizgilerin ve diğer grafiksel öğelerin renklerini çeşitli şekillerde tanımlayabiliriz. `'b'` mavi, `'g'` yeşil anlamına gelir ve bunun gibi MATLAB benzeri sözdizimleri kullanabiliriz. Çizgi stillerini seçmek için MATLAB API'sini de desteklenir: 

örneğin `'b.-'` noktalı mavi çizgi anlamına gelir:

```python
# MATLAB benzeri çizgi rengi ve stili
fig, ax = plt.subplots()
ax.plot(x, x**2, 'b.-') # noktalı mavi çizgi
ax.plot(x, x**3, 'g--') # yeşil kesikli çizgi
```

## Önerilen Yaklaşım: Argümanları kullanma

### color parametresi ile renk

Çizgi renklerini, adlarına veya RGB <a href='https://g.co/kgs/WEviGL'> hex kodlarına</a> göre de tanımlayabiliriz ve isteğe bağlı olarak `color` ve `alpha` argümanlarını kullanarak bunu sağlayabiliriz. Alpha çizginin opaklığını tanımlar

```python
fig, ax = plt.subplots()

ax.plot(x, x+1, color="magenta", alpha=0.2) # yarı şeffaf
ax.plot(x, x+2, color="#8B008B")         # RGB hex kodu
ax.plot(x, x+3, "#1f401c")         # RGB hex kodu
```

## Çizgi (line) ve işaretçi (marker) stilleri

## linewidth parametresi

Çizgi genişliğini değiştirmek için `linewidth` veya `lw` parametresini kullanabiliriz. 

```python
fig, ax = plt.subplots(figsize=(12,6))

# Linewidth veya lw kullanın
ax.plot(x, x-1, color="red", linewidth=0.25)
ax.plot(x, x-2, color="red", lw=0.50)
ax.plot(x, x-3, color="red", lw=1)
ax.plot(x, x-4, color="red", lw=10);
```

## Linestyles seçeneği

Aralarından seçim yapabileceğiniz birçok çizgi stili vardır.

```python
# olası çizgi stili seçenekleri: ‘--‘, ‘–’, ‘-.’, ‘:’
fig, ax = plt.subplots(figsize=(12,6))

ax.plot(x, x-1, color="green", lw=3, linestyle='-') # solid
ax.plot(x, x-2, color="blue", lw=3, ls='-.')        # dash ve dot
ax.plot(x, x-3, color="yellow", lw=3, ls=':')       # dots
ax.plot(x, x-4, color="red", lw=3, ls='--')         # dashes
```

### .set_dashes() ile özel Linestyle Dash dizisi tanımlama
Çizgi dizisi (dash), noktalardaki çizgiler ve boşlukların uzunluğunu tanımlayan eşit uzunluktaki bir kayan noktalar dizisidir. Örneğin, [5, 2, 1, 2] çizgi dizisi, 5 birim çizgi 2 birim boşluk ve 1 birim çizgi 2 birim boşluk tanımlar.

```python
fig, ax = plt.subplots(figsize=(12,6))

lines = ax.plot(x,x)

print(type(lines))
```

```python
fig, ax = plt.subplots(figsize=(12,6))
# özel dash
lines = ax.plot(x, x+8, color="black", lw=5)
lines[0].set_dashes([10, 10]) # format: 10 çizgi uzunluğu, 10 boşluk uzunluğu
```

```python
fig, ax = plt.subplots(figsize=(12,6))
# özel dash
lines = ax.plot(x, x+8, color="black", lw=5)
lines[0].set_dashes([1,1,1,1,10,10]) # format: çizgi uzunluğu, boşluk uzunluğu
```

## marker ve markersize parametreleri

Teknik olarak her zaman matplotlib veri noktaları arasına otomatik olarak bir çizgi çizer.

### Markers Style

İşaretleyici stillerinin listesi:https://matplotlib.org/stable/api/markers_api.html

```python
fig, ax = plt.subplots(figsize=(12,6))

# marker'ı string olarak tanımla
# markersize veya ms kullanarak integer boyut tanımla

ax.plot(x, x-1,marker='+',markersize=20)
ax.plot(x, x-2,marker='o',ms=20) # markersize veya ms kullanılabilir
ax.plot(x, x-3,marker='s',ms=20,lw=0) 
# yalnızca işaretçileri görmek için çizgi genişliğini sıfır yapın
ax.plot(x, x-4,marker='1',ms=20)
```

### markerfacecolor, markeredgewidth ve markeredgecolor

```python
fig, ax = plt.subplots(figsize=(12,6))

# marker boyutu ve rengi
ax.plot(x, x,color="black",lw=1, ls='-',marker='s',markersize=20, 
        markerfacecolor="red",markeredgewidth=8,markeredgecolor="blue");
```
