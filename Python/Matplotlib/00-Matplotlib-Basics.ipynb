# Basit Bir Örnek

### Basit Dizi Çizimi

**Veriyi Çizmek:**

```python
import numpy as np
```

```python
# Numpy'da 10 elemanlı ardışık tam sayı dizisi oluşturma
x = np.arange(0,10)
x
```

```python
# x'ten üretimiş bir fonksiyon dizisi
y = 2 * x
y
```

## Temel Matplotlib Komutları
Matplotlib plt.plot() fonksiyonu ile kolayca verilerinizi çizebilir.

Aşağıdaki komutları kullanarak çok basit bir çizgi grafiği oluşturabiliriz.

```python
# Fonksiyonel bir ilişkinin grafiği
plt.plot(x,y); # en basit matplotlib fonksiyon çizim çağrısı
```

```python
# çiziminize eksen ve başlık bilgisi ekleyebilirsiniz.
plt.plot(x,y) # en basit matplotlib fonksiyon çizim çağrısı
plt.xlabel('X Axis Title Here')
plt.ylabel('Y Axis Title Here')
plt.title('String Title Here')
```

### Daha fazla çizim parametresi düzenleme

```python
plt.plot(x,y) # en basit matplotlib fonksiyon çizim çağrısı
plt.xlabel('X Axis Title Here')
plt.ylabel('Y Axis Title Here')
plt.title('String Title Here')
plt.xlim(0,6)  # X ekseni --> (Alt Limit, Üst Limit)
plt.ylim(0,12) # Y ekseni --> (Alt Limit, Üst Limit)
```

### subplot ile çoklu çizim

```python
plt.subplot(121)

plt.subplot(122)
```

```python
plt.subplot(121)
plt.plot(x,x ** 2,'r')

plt.subplot(122)
plt.plot(x,np.sin(x),'cyan')
plt.tight_layout()
```
